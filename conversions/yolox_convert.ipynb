```python
import torch

from yolox.exp.default.yolox_nano import Exp
import iree.turbine.aot as aot
```

```python
exp = Exp()
model = exp.get_model().cpu()
model.load_state_dict(
    torch.load(
        "../models/yolox_nano.pth", map_location=torch.device("cpu"), weights_only=False
    )["model"],
)

model = model.eval()
sample_inputs = torch.randn((1, 3, 416, 416))
model.head.decode_in_inference = False

print(f"{model.forward(sample_inputs).shape}")


export_output = aot.export(model, sample_inputs)

export_output.save_mlir("converted_yoloxn_config.mlir")
```
