```python
class Counter(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.counter = torch.nn.Parameter(
            torch.tensor([0], dtype=torch.int32), requires_grad=False
        )

    def get_counter(self):
        return self.counter

    def set_counter(self, c):
        self.counter.data = c

    def add_counter(self, c):
        self.counter.data = c + self.counter.data


model = Counter()


model = model.eval()

print(f"{model}")

print(model.get_counter())
model.set_counter(torch.tensor([12], dtype=torch.int32, requires_grad=False))
print(model.get_counter())
model.add_counter(torch.tensor([15], dtype=torch.int32, requires_grad=False))
print(model.get_counter())

# export_output = aot.export(model, dynamic_shapes=)
# compiled_model = torch.compile(model, backend="iree_turbine")
```
